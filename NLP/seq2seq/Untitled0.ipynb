```python
import torch
import numpy as np
```

```python
a = torch.tensor(np.arange(36).reshape((4, 3, 3)))
a
```

```python
mask = torch.tensor([[1, 1, 1], [1, 1, 0], [1, 0, 0], [1, 0, 0]])
```

```python
tmp = a * mask.unsqueeze(-1)
print(tmp[:, :, 0].unsqueeze(-1))
tmp.to(torch.float).mean(-1)
```

```python
probs = torch.nn.Softmax(dim=1)(tmp[:, :, 0].unsqueeze(-1).to(torch.float))
probs
```

```python
print(tmp)
(tmp * probs).sum(axis=1)
```

```python
print(torch.nn.Softmax()(torch.tensor([10, 13, 0]).to(torch.float)))
torch.nn.Softmax()(torch.tensor([11, 14, 0]).to(torch.float))
```
